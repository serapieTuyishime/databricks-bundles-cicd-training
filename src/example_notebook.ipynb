```python
# Create a simple DataFrame and save it as a Delta table
from pyspark.sql import SparkSession
from datetime import datetime

spark = SparkSession.builder.getOrCreate()

# Create sample data
data = [
    (1, "example_row_1", datetime.now()),
    (2, "example_row_2", datetime.now()),
    (3, "example_row_3", datetime.now())
]

df = spark.createDataFrame(data, ["id", "name", "timestamp"])

# Write to Delta table
df.write.format("delta").mode("overwrite").saveAsTable("default.example_output")

print(f"Successfully wrote {df.count()} rows to default.example_output")
```
