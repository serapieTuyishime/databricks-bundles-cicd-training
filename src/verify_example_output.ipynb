```python
from pyspark.sql.utils import AnalysisException

catalog = spark.sql("SELECT current_catalog()").collect()[0][0]
schema = spark.sql("SELECT current_schema()").collect()[0][0]

table_name = f"{catalog}.{schema}.example_output"

print("Verifying table:", table_name)

try:
    spark.table(table_name)
    print("VERIFICATION SUCCESS: table exists")
except AnalysisException:
    raise Exception(f"VERIFICATION FAILED: table {table_name} does not exist")
```
